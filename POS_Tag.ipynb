```python
paragraph = """Sunset is the time of day when our sky meets the outer space solar winds. There are blue, pink, and purple swirls, spinning and twisting, like clouds of balloons caught in a whirlwind. The sun moves slowly to hide behind the line of horizon, while the moon races to take its place in prominence atop the night sky."""
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk.corpus import stopwords
sentences = nltk.sent_tokenize(paragraph)
```

```python
for i in range (len(sentences)):

 words =nltk.word_tokenize(sentences[i])
 words =[word for word in words if word not in set (stopwords.words('english'))]
 pos_tag = nltk.pos_tag(words)
 print(pos_tag)
```
