```python
import opendatasets as od
od.download('https://www.kaggle.com/datasets/saurabhshahane/ecommerce-text-classification')
```

```python
import pandas as pd
df = pd.read_csv('/content/ecommerce-text-classification/ecommerceDataset.csv')
```

```python
df.head(3)
```

```python
df.Household.value_counts()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.Household.replace('Clothing & Accessories', 'Clothing_Accessories',inplace=True)
df.Household.unique()
```

```python
df['Household'] = '__label__' +df['Household'].astype(str)
df.head(3)
```

```python
original_description_col_name = df.columns[1]
df['Household_description'] = df['Household'] + ' ' + df[original_description_col_name].apply(preprocess)
df.head(3)
```

```python
df['Household_description'][0]
```

```python
import re
def preprocess(text):
    text = str(text) # Ensure the input is a string
    text = text.lower() # Convert to lowercase
    text = re.sub(r'[^ء-ي٠-٩A-Za-z0-9 ]', ' ', text) # Remove special characters, keeping only Arabic, English alphanumeric and spaces
    text = re.sub(r'\s+', ' ', text).strip() # Replace multiple spaces with a single one and strip leading/trailing spaces
    return text
```

```python
preprocess('VIKI Bookcase/Bookshelf (3-shelf/shelve, Whitee) |? .hi ')
```

```python
df.head(4)
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2)
```

```python
train.shape, test.shape
```

```python
test.head(3)
```

```python
train.to_csv('ecommerce.train', columns=['Household_description'], index=False, header=False)
train.to_csv('ecommerce.test', columns=['Household_description'], index=False, header=False)
```

```python
import fasttext
model = fasttext.train_supervised(input='ecommerce.train')
```

```python
model.test('ecommerce.test')
```
